# Tech

Here's how you get a tech job bru.

## Coding Practice

- HackerRank: https://www.hackerrank.com
    - I like this because there are discussion forums and companies on here. It's nice to also get skills recognized and gain stars - easy/medium for quant sellside and medium (maybe some hards) for tech/quant buyside

- Leetcode: https://leetcode.com
    - Good problems and premium is cheap for what you get. Honestly as a quant/data scientist, I see no reason to be on here with the CS kids unless you want to to SWE or a more code focused ML role like ML engineers.

- https://coderpad.io
    - During covid, lots of companies use coderpad to see you code real time during final rounds. I know Goldman and some tech companies like Facebook use it frequently. You can't practice many problems, but you can get a free trial and try the sandbox version
    
- https://github.com/tayllan/awesome-algorithms
    - This is a MASSIVE collection of algorithms if you really want to sharpen your coding skills. IMO, not much time to do so if you're planning on recruiting in the spring, but this will make you much better at CS fundamentals.
    
- https://www.kaggle.com
    - This is a really cool way to learn data science through practice. They have notebooks for you to follow how others parsed through datasets and messy datasets for you to wrangle with. This is a good resume bullet if you don't already have data science experience.
